## SNOMED-CT codes to OHDSI ATLAS
 - Use these snomed_ct codes (6,327) to OHDSI atlas. Only 6,326 SNOMED-CT codes mapped and 1 snomed-ct codes
   (107360810001191) did not mapped to OHDSI ATLAS.
 - Now we have chosen Vocabulary = SNOMED (6,326); and then selected descendent concepts which brought 27,752 total concepts 
   Which contains ICDO3, SNOMED, Nebraska Lexicon etc. We have chosen Vocabulary = SNOMED-CT (12,569).
 - OHDSI Atlas data queried on 02-05-2025 (https://atlas-demo.ohdsi.org/#/conceptset/0/included). Finally 12,569 SNOMED-CT codes acquired.
 

```python
# Following txt file contains all rare diseases with its descendants IDs
rd_descendants = pd.read_csv('SNOMED_CT_with_Descendants_from_OHDSI.txt', delimiter='\t')
# Drop the first null column 
rd_descendants = rd_descendants.drop(rd_descendants.columns[0], axis=1)
rd_descendants
```

```python
rd_descendants = rd_descendants[['Code','Name']]
rd_descendants
```

```python
## INvestigate rare disease with GARD Ids are existing in OHDSI list or not. Here one disease is missing because no mapping at OHDSI. 
len(set(GARD_Orpha_snomed_final['SNOMED concept ID'].tolist()) & set(rd_descendants['Code'].tolist()))
```

```python
import requests
import json
from urllib.parse import quote

def get_hpo(dis):
    # Remove any null bytes or unwanted characters from the input term
    dis = dis.replace('\x00', '')  # Remove any null characters
    # URL encode the disease term to handle spaces and special characters correctly
    encoded_dis = quote(dis)
    url = f'https://clinicaltables.nlm.nih.gov/api/hpo/v3/search?terms={encoded_dis}'
    print(url)  # Print the URL for debugging
    my_response = requests.get(url)
    return my_response

# Specify the output txt file name
output_file_name = "C:\\Users\\yadawa2\\Desktop\\RD_Cohort_Paper_Feb_2025\\SNOMED_CT_Code_mapping\\SNOMED_CT_codes_with_HPO_tag.txt"

# Open the output file in write mode with UTF-16 encoding
with open(output_file_name, 'w', encoding='utf-16-le', newline='') as output_file:
    # Write the header row (Code, Name, HPO)
    output_file.write("Code\tName\tHPO\n")
    
    # Iterate through each row in the DataFrame
    for index, row in rd_descendants.iterrows():
        # Extract the Code and Name from the row
        code = row['Code']
        name = row['Name']
        
        # Ensure the disease name is properly handled for spaces and special characters
        print(f"Code: {code}, Name: {name}")  # Print for debugging
        
        # Call the get_hpo function to retrieve information
        info = get_hpo(name)
        
        if info.ok:
            j_data = json.loads(info.content)
            # For simplicity, you can convert the JSON data to a string (you could format it better if needed)
            hpo_data = json.dumps(j_data)
            # Write the Code, Name, and HPO data to the output txt file
            output_file.write(f"{code}\t{name}\t{hpo_data}\n")
        else:
            # Handle case where the response isn't OK (if necessary)
            print(f"Failed to retrieve data for {name}")
```
